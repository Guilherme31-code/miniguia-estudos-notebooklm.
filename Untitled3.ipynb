# Carregando e Processando Dados

```python
from google.colab import drive
import os

caminho_dados = "/content/drive/MyDrive/Dados /2023_Viagem.csv"

# Mount Google Drive if not already mounted
if not os.path.exists('/content/drive'):
    print("Montando Google Drive...")
    drive.mount('/content/drive')
else:
    print("Google Drive já está montado.")

# Check if the file exists
if os.path.exists(caminho_dados):
    print(f"O arquivo '{os.path.basename(caminho_dados)}' foi encontrado no Google Drive.")
else:
    print(f"O arquivo '{os.path.basename(caminho_dados)}' NÃO foi encontrado no Google Drive no caminho: {caminho_dados}")
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)

try:
    df_viagens = pd.read_csv(caminho_dados, encoding='latin1', sep=';')
    print("Dados carregados com sucesso!")
except FileNotFoundError:
    print(f"Erro: O arquivo não foi encontrado em {caminho_dados}. Verifique o caminho.")
except Exception as e:
    print(f"Ocorreu um erro ao carregar o arquivo: {e}")
```

```python
try:
    # Substituir vírgulas por pontos e converter para float
    df_viagens['Valor diárias'] = df_viagens['Valor diárias'].astype(str).str.replace(',', '.', regex=False).astype(float)
    df_viagens['Valor passagens'] = df_viagens['Valor passagens'].astype(str).str.replace(',', '.', regex=False).astype(float)

    # Calcular 'Valor Total'
    df_viagens['Valor Total'] = df_viagens['Valor diárias'] + df_viagens['Valor passagens']

    print("Colunas 'Valor diárias', 'Valor passagens' e 'Valor Total' processadas com sucesso!")
    display(df_viagens[['Valor diárias', 'Valor passagens', 'Valor Total']].head())
except KeyError as e:
    print(f"Erro: Coluna não encontrada. Verifique se as colunas 'Valor diárias' e 'Valor passagens' existem: {e}")
except ValueError as e:
    print(f"Erro de conversão: Verifique se os dados nas colunas 'Valor diárias' e 'Valor passagens' podem ser convertidos para números: {e}")
except Exception as e:
    print(f"Ocorreu um erro inesperado durante o processamento: {e}")
```

```python
gastos_totais_por_cargo = df_viagens.groupby('Cargo')['Valor Total'].max().reset_index().sort_values(by='Valor Total', ascending=False)
```

```python
gastos_totais_por_cargo[gastos_totais_por_cargo['Valor Total'] > 1000000]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort and get the top 10 cargos by 'Valor Total'
top_10_cargos = gastos_totais_por_cargo.head(10)

# Create the bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x='Valor Total', y='Cargo', data=top_10_cargos, hue='Cargo', palette='viridis', legend=False)
plt.title('Top 10 Cargos com Maiores Gastos Totais')
plt.xlabel('Valor Total (R$)')
plt.ylabel('Cargo')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
